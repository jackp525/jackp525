# Data Dictionary

```python
#Data Dictonary

#p1                   peak electricity rate between 0-100 hours ($0.110)
#p2                   peak electricity rate between 101-250 hours ($0.180)
#p3                   peak electricity rate of 251 hours or more ($0.230)

#mp1                  mid-peak electricity rate between 0-100 hours ($0.081)
#mp2                  mid-peak electricity rate between 101-250 hours ($0.127)
#mp3                  mid-peak electricity rate of 251 hours or more ($0.160)

#op1                  off-peak electricity rate between 0-100 hours ($0.038)
#op2                  off-peak electricity rate between 101-250 hours ($0.064)
#op3                  off-peak electricity rate of 251 hours or more ($0.120)

#Tran_charge          transmission charge(1.5%)
#Demand_Charge        demand side management charge (8.5%)
#Tax_charge           tax charge (8.0%)

#last_name            last name on account
#account_number       account number
#off_peak             off-peak hours for account
#mid_peak             mid-peak hours for account
#peak                 peak hours for account

#sum1                 sum of off-peak usage charge
#sum2                 sum of mid-peak usage charge
#sum3                 sum of peak usage charge

#lowbill              usage charge for hours between 0-100
#midbill              usage charge for hours between 101-250
#highbill             usage charge for above 251

#total_uc             total usage charge (sum123)

#transmission_charge  total of transmission charge
#managment_charge     total of demand side management charge
#tax_charge           total of tax charge
#sumtotal             total of usage charge + transmission + tax + demand

#range1_max           range 0-100 max (100)
#range2_min           range 101-250 min (101)
#range2_max           range 101-250 max (250)
#range2_hours         number of hours in range 101-250 inclusive (150)
```

# Constants

```python
#Electricity Bill Constants

p1= .11 #0-100hr peak
p2= .18 #101-250hr peak
p3= .23 #251+hr peak

mp1= 0.081 #0-100hr peak
mp2= 0.127 #101-250hr peak
mp3= 0.160 #251+hr peak

op1= 0.038 #0-100hr peak
op2= 0.064 #101-250hr peak
op3= 0.120 #251+hr peak

Tran_charge= 0.015 #transmission charge
Demand_charge= 0.085 #demand side management charge
Tax_charge= 0.08 #tax charge

minimum_charge= 10 #minimum electricity bill charge

range1_max= 100 #range 0-100 max (100)
range2_min= 101 #range 101-250 min (101)
range2_max= 250 #range 101-250 max (250)
range2_hours= 150 #number of hours in range 101-250 inclusive
```

# Inputs

```python
#prompt user for last name, account#, peak hours, mid-peak hours, off-peak hours
print('Input Electricity Bill Account information:')
print('\n')

print('Last name:')
last_name= input()
print('\n')

while True:
  try:
    print('Account number:')
    account_number= int(input())
    print('\n')
  except ValueError:
    print("Invalid account number. Please input an integer.")
    print('\n')
  else:
    break

print('\n')

print('Off-Peak hours:')
off_peak= int(input())
print('\n')

print('Mid-Peak hours:')
mid_peak= int(input())
print('\n')

print('Peak hours:')
peak= int(input())
```

# Calculations Usage Charge:

```python
#off-peak hours calculations

print('Off-Peak Energy Use (Kwh):', off_peak, )
if off_peak<= range1_max:
  sum1= op1* off_peak
  print("Off-Peak Energy Usage:", sum1)

elif range2_min <= off_peak <= range2_max:
  lowbill= op1 * range1_max
  Extra_hr= off_peak - range1_max
  midbill= op2 * Extra_hr
  sum1= lowbill + midbill
  print("Off-Peak Energy Usage:", sum1)

else:
  lowbill= op1 * range1_max
  midbill= op2 * range2_hours
  highbill= (off_peak-range2_max)* op3
  sum1= lowbill + midbill + highbill
  print("Off-Peak Energy Usage:", sum1)

#mid-peak hours calculations

print('\n')
print('Mid-Peak Energy Use (Kwh):', mid_peak, )
if mid_peak<= range1_max:
  sum2= mp1* mid_peak
  print("Mid-Peak Energy Usage:", sum2)

elif range2_min <= mid_peak <= range2_max:
  lowbill= mp1 * range1_max
  Extra_hr= mid_peak - range1_max
  midbill= mp2 * Extra_hr
  sum2= lowbill + midbill
  print("Mid-Peak Energy Usage:", sum2)

else:
  lowbill= mp1 * range1_max
  midbill= mp2 * range2_hours
  highbill= (mid_peak-range2_max)* mp3
  sum2= lowbill + midbill + highbill
  print("Mid-Peak Energy Usage:", sum2)

#peak hours calculations

print('\n')
print('Peak Energy Use (Kwh):', peak, )
if peak<= range1_max:
  sum3= p1* peak
  print("Peak Energy Usage:", sum3)

elif range2_min <= peak <= range2_max:
  lowbill= p1 * range1_max
  Extra_hr= peak - range1_max
  midbill= p2 * Extra_hr
  sum3= lowbill + midbill
  print("Peak Energy Usage:", sum3)

else:
  lowbill= p1 * range1_max
  midbill= p2 * range2_hours
  highbill= (peak-range2_max)* p3
  sum3= lowbill + midbill + highbill
  print("Peak Energy Usage:", sum3)
```

# Total Usage Charge:

```python
#total usage charge
total_uc = sum1 + sum2 + sum3
print("Total Usage Charge:", total_uc)
```

# Accounting For Total Usage less than $10
(Calculating Fees and Grand total)

```python
if total_uc < minimum_charge:
  total_uc = minimum_charge
  #transmission charge calculation
  transmission_charge= Tran_charge * total_uc
  print('Transmission Charge (1.5% of Usage Charge):', transmission_charge)
  #Demand Side Management Charge
  managment_charge= Demand_charge * total_uc
  print('Demand Side Management Charge (8.5% of Usage Charge):', managment_charge)
  #Tax Charge
  tax_charge= Tax_charge * total_uc
  print('Tax Charge (8%):', tax_charge)
  sumtotal= total_uc + transmission_charge + managment_charge + tax_charge
  print('Total Electrical Charge:', sumtotal)
else:
  total_uc = total_uc
   #transmission charge calculation
  transmission_charge= Tran_charge * total_uc
  print('Transmission Charge (1.5% of Usage Charge):', transmission_charge)
  #Demand Side Management Charge
  managment_charge= Demand_charge * total_uc
  print('Demand Side Management Charge (8.5% of Usage Charge):', managment_charge)
  #Tax Charge
  tax_charge= Tax_charge * total_uc
  print('Tax Charge (8%):', tax_charge)
  sumtotal= total_uc + transmission_charge + managment_charge + tax_charge
  print('Total Electrical Charge:', sumtotal)
```

# Formatting Electricity Bill

```python
print('|'+('-'*75)+'|') #table line
print(f'|\033[1m{"Mile High Energy Inc.":^75}\033[0m|') #title

print(f'|\033[1m{"Denver, CO 80022":^75}\033[0m|') #address
print('|'+('-'*75)+'|') #table line

print(f'|{"Account Owner":>15}{(last_name):>60}|') #account name
#print(f'{(last_name):>75}') # last name input

print('|'+('-'*75)+'|') #table line

print(f'|{"Account Number":>16}{(account_number):>59}|') #account number

print('|'+('-'*75)+'|') #table line

print(f'|{"Off-Peak Energy Use (Kwh)":>27}{off_peak:>48}|') #Off-Peak Energy Use (Kwh)


print('|'+('-'*75)+'|') #table line

print(f'|{"Off-Peak Charge":>17}{"${:,.2f}".format(sum1):>58}|') #Off-Peak Charge

print('|'+('-'*75)+'|') #table line

print(f'|{"Mid-Peak Energy Use (Kwh)":>27}{mid_peak:>48}|') #Mid-Peak Energy Use (Kwh)

print('|'+('-'*75)+'|') #table line

print(f'|{"Mid-Peak Charge":>17}{"${:,.2f}".format(sum2):>58}|') #Mid-Peak Charge

print(('|'+'-'*75)+'|') #table line

print(f'|{"Peak Energy Use (Kwh)":>23}{peak:>52}|') #Peak Energy Use (Kwh)


print('|'+('-'*75)+'|') #table line

print(f'|{"Peak Charge":>13}{"${:,.2f}".format(sum3):>62}|') #Peak Charge


print('|'+('-'*75)+'|') #table line

print(f'|{"Total Usage Charge":>20}{"${:,.2f}".format(total_uc):>55}|') #total usage Charge

print('|'+('-'*75)+'|') #table line

print(f'|{"Transmission Charge (1.5% of Usage Charge)":>44}{"${:,.2f}".format(transmission_charge):>31}|') #transmission charge

print('|'+('-'*75)+'|') #table line

print(f'|{"Demand Side Management Charge (8.5% of Usage Charge)":>54}{"${:,.2f}".format(managment_charge):>21}|') #management charge

print('|'+('-'*75)+'|') #table line

print(f'|{"Tax Charge (8%)":>17}{"${:,.2f}".format(tax_charge):>58}|') #tax charge

print('|'+('-'*75)+'|') #table line

print(f'|{"Total Electrical Charge":>25}{"${:,.2f}".format(sumtotal):>50}|') #total usage Charge


print('|'+('-'*75)+'|') #table line
print(f'|{"Thank You":^75}|')
print('|'+('-'*75)+'|') #table line
```

```python
#prompt user for last name, account#, peak hours, mid-peak hours, off-peak hours
print('Input Electricity Bill Account information:')
print('\n')

print('Last name:')
last_name= input()
print('\n')

while True:
  try:
    print('Account number:')
    account_number= int(input())
    print('\n')
  except ValueError:
    print("Invalid account number. Please input an integer.")
    print('\n')
  else:
    break

print('\n')

print('Off-Peak hours:')
off_peak= int(input())
print('\n')

print('Mid-Peak hours:')
mid_peak= int(input())
print('\n')

print('Peak hours:')
peak= int(input())

#off-peak hours calculations

print('Off-Peak Energy Use (Kwh):', off_peak, )
if off_peak<= range1_max:
  sum1= op1* off_peak
  print("Off-Peak Energy Usage:", sum1)

elif range2_min <= off_peak <= range2_max:
  lowbill= op1 * range1_max
  Extra_hr= off_peak - range1_max
  midbill= op2 * Extra_hr
  sum1= lowbill + midbill
  print("Off-Peak Energy Usage:", sum1)

else:
  lowbill= op1 * range1_max
  midbill= op2 * range2_hours
  highbill= (off_peak-range2_max)* op3
  sum1= lowbill + midbill + highbill
  print("Off-Peak Energy Usage:", sum1)

#mid-peak hours calculations

print('\n')
print('Mid-Peak Energy Use (Kwh):', mid_peak, )
if mid_peak<= range1_max:
  sum2= mp1* mid_peak
  print("Mid-Peak Energy Usage:", sum2)

elif range2_min <= mid_peak <= range2_max:
  lowbill= mp1 * range1_max
  Extra_hr= mid_peak - range1_max
  midbill= mp2 * Extra_hr
  sum2= lowbill + midbill
  print("Mid-Peak Energy Usage:", sum2)

else:
  lowbill= mp1 * range1_max
  midbill= mp2 * range2_hours
  highbill= (mid_peak-range2_max)* mp3
  sum2= lowbill + midbill + highbill
  print("Mid-Peak Energy Usage:", sum2)

#peak hours calculations

print('\n')
print('Peak Energy Use (Kwh):', peak, )
if peak<= range1_max:
  sum3= p1* peak
  print("Peak Energy Usage:", sum3)

elif range2_min <= peak <= range2_max:
  lowbill= p1 * range1_max
  Extra_hr= peak - range1_max
  midbill= p2 * Extra_hr
  sum3= lowbill + midbill
  print("Peak Energy Usage:", sum3)

else:
  lowbill= p1 * range1_max
  midbill= p2 * range2_hours
  highbill= (peak-range2_max)* p3
  sum3= lowbill + midbill + highbill
  print("Peak Energy Usage:", sum3)

#total usage charge
total_uc = sum1 + sum2 + sum3
print("Total Usage Charge:", total_uc)

#total calc
if total_uc < minimum_charge:
  total_uc = minimum_charge
  #transmission charge calculation
  transmission_charge= Tran_charge * total_uc
  print('Transmission Charge (1.5% of Usage Charge):', transmission_charge)
  #Demand Side Management Charge
  managment_charge= Demand_charge * total_uc
  print('Demand Side Management Charge (8.5% of Usage Charge):', managment_charge)
  #Tax Charge
  tax_charge= Tax_charge * total_uc
  print('Tax Charge (8%):', tax_charge)
  sumtotal= total_uc + transmission_charge + managment_charge + tax_charge
  print('Total Electrical Charge:', sumtotal)
else:
  total_uc = total_uc
   #transmission charge calculation
  transmission_charge= Tran_charge * total_uc
  print('Transmission Charge (1.5% of Usage Charge):', transmission_charge)
  #Demand Side Management Charge
  managment_charge= Demand_charge * total_uc
  print('Demand Side Management Charge (8.5% of Usage Charge):', managment_charge)
  #Tax Charge
  tax_charge= Tax_charge * total_uc
  print('Tax Charge (8%):', tax_charge)
  sumtotal= total_uc + transmission_charge + managment_charge + tax_charge
  print('Total Electrical Charge:', sumtotal)

#formatting code receipt
print('|'+('-'*75)+'|') #table line
print(f'|\033[1m{"Mile High Energy Inc.":^75}\033[0m|') #title

print(f'|\033[1m{"Denver, CO 80022":^75}\033[0m|') #address
print('|'+('-'*75)+'|') #table line

print(f'|{"Account Owner":>15}{(last_name):>60}|') #account name
#print(f'{(last_name):>75}') # last name input

print('|'+('-'*75)+'|') #table line

print(f'|{"Account Number":>16}{(account_number):>59}|') #account number

print('|'+('-'*75)+'|') #table line

print(f'|{"Off-Peak Energy Use (Kwh)":>27}{off_peak:>48}|') #Off-Peak Energy Use (Kwh)


print('|'+('-'*75)+'|') #table line

print(f'|{"Off-Peak Charge":>17}{"${:,.2f}".format(sum1):>58}|') #Off-Peak Charge

print('|'+('-'*75)+'|') #table line

print(f'|{"Mid-Peak Energy Use (Kwh)":>27}{mid_peak:>48}|') #Mid-Peak Energy Use (Kwh)

print('|'+('-'*75)+'|') #table line

print(f'|{"Mid-Peak Charge":>17}{"${:,.2f}".format(sum2):>58}|') #Mid-Peak Charge

print(('|'+'-'*75)+'|') #table line

print(f'|{"Peak Energy Use (Kwh)":>23}{peak:>52}|') #Peak Energy Use (Kwh)


print('|'+('-'*75)+'|') #table line

print(f'|{"Peak Charge":>13}{"${:,.2f}".format(sum3):>62}|') #Peak Charge


print('|'+('-'*75)+'|') #table line

print(f'|{"Total Usage Charge":>20}{"${:,.2f}".format(total_uc):>55}|') #total usage Charge

print('|'+('-'*75)+'|') #table line

print(f'|{"Transmission Charge (1.5% of Usage Charge)":>44}{"${:,.2f}".format(transmission_charge):>31}|') #transmission charge

print('|'+('-'*75)+'|') #table line

print(f'|{"Demand Side Management Charge (8.5% of Usage Charge)":>54}{"${:,.2f}".format(managment_charge):>21}|') #management charge

print('|'+('-'*75)+'|') #table line

print(f'|{"Tax Charge (8%)":>17}{"${:,.2f}".format(tax_charge):>58}|') #tax charge

print('|'+('-'*75)+'|') #table line

print(f'|{"Total Electrical Charge":>25}{"${:,.2f}".format(sumtotal):>50}|') #total usage Charge


print('|'+('-'*75)+'|') #table line
print(f'|{"Thank You":^75}|')
print('|'+('-'*75)+'|') #table line
```
